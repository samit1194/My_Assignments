like Python's built-in function filter()

```python
'''
my_list = [1,3,5,7]

final_list = []

f = lambda x : x*2 

for i in my_list:
    final_list.append(f(i))
    
final_list
'''
```

```python
def myfilter(func,my_list):
    final_list = [] 
    for element in my_list : 
        final_list.append(func(element))
    return(final_list)
```

```python
# multiplied by 2

f = lambda x : x*2
inp1 = [1,3,5,7,9]
print("myfilter : ",myfilter(f, inp1))
#print("filter : ",list(filter(f , inp1)))
```

```python
# Cubes of integers

f = lambda x : x**2
inp1 = [1,3,5,7,9]
print("myfilter : ",myfilter(f, inp1))
#print("filter : ",list(filter(f , inp1)))
```
